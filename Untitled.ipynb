## Choosing the Value of K for K-Mean Clustering

### Silhouette Score for K-mean


$$\text{silhouette score}=\frac{p-q}{max(p,q)}$$

  is the mean distance to the points in the nearest cluster that the data point is not a part of

  is the mean intra-cluster distance to all the points in its own cluster.

* The value of the silhouette score range lies between -1 to 1.

* A score closer to 1 indicates that the data point is very similar to other data points in the cluster,

* A score closer to -1 indicates that the data point is not similar to the data points in its cluster.


```python
from sklearn.cluster import KMeans
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_coefficient=[]

for k in range(2,10):
    # intialise kmeans
    kmeans=KMeans(n_clusters=k,max_iter=50)
    kmeans.fit(scaled_df)
    
    # silhouette score
    silhouette_avg = silhouette_score(scaled_df,kmeans.labels_)
    silhouette_coefficient.append(silhouette_avg)
silhouette_coefficient
```

```python
plt.title('silhouette_score\n')
plt.plot(range(2,10),silhouette_coefficient)
plt.xlabel('n_clusters')
plt.ylabel('silhouette_coefficient')
plt.show()
```

### Plotting Elbow Curve

```python
sse=[]
for k in range (2,8):
    kmeans=KMeans(n_clusters=k,max_iter=50)
    kmeans.fit(scaled_df)
    sse.append(kmeans.inertia_)
sse
```

```python
plt.title('elbow curve')
plt.plot(range(2,8),sse)
plt.xlabel('n_clusters')
plt.ylabel('sse')
plt.show()
```

#### So, from the Silhouette score curve & Elbow curve we can conclue that k=3 is significant

```python
# final model with k=3
kmeans = KMeans(n_clusters=3, max_iter=50)
kmeans.fit(scaled_df)
```

```python
kmeans.labels_
```

```python
# assign the label
df['cluster_id'] = kmeans.labels_
df.head()
```

## Hierarchical Clustering

```python
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import cut_tree
```

```python
scaled_df.head()
```

```python
df.head()
```

```python
# single linkage
plt.figure(figsize=(10,6))

mergings = linkage(scaled_df, method="single", metric='euclidean')
dendrogram(mergings)

plt.title('Hierarchical Clustering Dendrogram - single linkage',fontsize=20)
plt.xlabel('Country',fontsize=20)
plt.show()
```

It's not clearly visible in single linkage dendogram hence we will go for complete linkage which gives us proper results.

```python
# complete linkage
plt.figure(figsize=(10,6))

mergings = linkage(scaled_df, method="complete", metric='euclidean')
dendrogram(mergings)

plt.title('Hierarchical Clustering Dendrogram - complete linkage',fontsize=20)
plt.xlabel('Country',fontsize=20)
plt.show()
```

Let us cut the dendrogram vertically such that n_clusters = 4

```python
# 3 clusters
cluster_labels = cut_tree(mergings, n_clusters=4).reshape(-1, )
cluster_labels
```

```python
# assign cluster labels
df['cluster_labels'] = cluster_labels
df.head()
```

```python
plt.figure(figsize=(26, 30))
sns.set_style('whitegrid')
sns.set_context('talk')

plt.subplot(4,3,1)
sns.boxplot(x='cluster_labels',y='child_mort',palette='husl', data = df)
plt.title('Child Mortality',fontsize=20)

plt.subplot(4,3,2)
sns.boxplot(x='cluster_labels',y ='exports',palette='husl', data = df)
plt.title('Exports',fontsize=20)

plt.subplot(4,3,3)
sns.boxplot(x='cluster_labels',y = 'gdpp', palette='husl',data = df)
plt.title('GDPP',fontsize=20)

plt.subplot(4,3,4)
sns.boxplot(x='cluster_labels',y = 'health',palette='husl', data = df)
plt.title('Health',fontsize=20)

plt.subplot(4,3,5)
sns.boxplot(x='cluster_labels',y = 'imports',palette='husl',  data = df)
plt.title('Imports',fontsize=20)

plt.subplot(4,3,6)
sns.boxplot(x='cluster_labels',y= 'income',palette='husl', data = df)
plt.title('Income',fontsize=20)

plt.subplot(4,3,7)
sns.boxplot(x='cluster_labels',y = 'inflation',palette='husl', data = df)
plt.title('Inflation',fontsize=20)

plt.subplot(4,3,8)
sns.boxplot(x='cluster_labels',y= 'life_expec',palette='husl', data = df)
plt.title('Life Expectancy',fontsize=20)

plt.subplot(4,3,9)
sns.boxplot(x='cluster_labels',y= 'total_fer',palette='husl', data = df)
plt.title('Total Fertility',fontsize=20)

plt.show()
```
